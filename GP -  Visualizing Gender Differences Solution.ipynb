## Guided Project: Visualizing The Gender Gap In College Degrees

#### Library Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Enable Inline Visualizations

```python
%matplotlib inline
```

#### Load File

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

#### Create Color Palette

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

#### dataquest.io Given Degree Lists

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

#### Create a List of Lists

```python
cats = [stem_cats, lib_arts_cats, other_cats]
```

#### Loop Through Outer List and Each Inner List in it

```python
fig = plt.figure(figsize=(15, 18))

for outer_counter, outer_list_name in enumerate(cats): #Elements of outer list are the three inner lists of degree categories
    for inner_counter, inner_list_value in enumerate(cats[outer_counter]): #Elements of each of the three inner list are
                                                                     #the degree descriptions
        ax = fig.add_subplot(6, 3, outer_counter + (inner_counter * 3) + 1)
        ax.plot(women_degrees['Year'], women_degrees[cats[outer_counter][inner_counter]], 
                c = cb_dark_blue, label = 'Women', linewidth = 3)
        ax.plot(women_degrees['Year'], 100-women_degrees[cats[outer_counter][inner_counter]],
                c = cb_orange, label='Men', linewidth = 3)
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_yticks([0,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        ax.set_title(cats[outer_counter][inner_counter])
        ax.tick_params(bottom = False, top = False, left = False, right = False, labelbottom=False)
        
        if (outer_counter == 0) and (inner_counter == 0):
            ax.text(2003, 87, 'Women')
            ax.text(2005, 8, 'Men')
        elif (outer_counter == 0) and (inner_counter == 5):
            ax.text(2003, 6, 'Women')
            ax.text(2005, 90, 'Men')
            ax.tick_params(labelbottom=True)
        elif (outer_counter == 1) and (inner_counter == 0):
            ax.text(2003, 80, 'Women')
            ax.text(2005, 15, 'Men')
        elif (outer_counter == 1) and (inner_counter == 4):
            ax.tick_params(labelbottom=True)
        elif (outer_counter == 2) and (inner_counter == 0):
            ax.text(2003, 75, 'Women')
            ax.text(2005, 25, 'Men')
        elif (outer_counter == 2) and (inner_counter == 5):
            ax.text(2003, 65, 'Women')
            ax.text(2005, 30, 'Men')
            ax.tick_params(labelbottom=True)

#plt.savefig('gender_degrees.png') #ENABLE to save as a png file
plt.show()
```
